```python
import pandas as pd
import os

# Find the line numbers in the text file associated with the task

def findlines(filename):
    t_starts=[]
    with open(filename) as file:
        for linenum, linetxt in enumerate(file,1):
            if "Task started at" in linetxt:
                t_starts.append(linenum)
    return (t_starts)

# Currently inputting name of text file but will have it loop through
fname='sub-1003_ses-Ex_task-MST_run-1.txt'
t_starts=findlines(fname)
test_len = 140


test_data=pd.read_csv(fname,sep=',',skiprows=t_starts[0],header=0,nrows=test_len)
Stim_file=test_data[['Stim','Cond','StartT','Resp','Corr','RT']]
```
