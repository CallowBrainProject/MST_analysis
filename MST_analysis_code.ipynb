```python
import pandas as pd
import os


def findlines(filename):
    s_starts=[]
    t_starts=[]
    with open(filename) as file:
        for linenum, linetxt in enumerate(file,1):
            if "Study phase started at" in linetxt:
                s_starts.append(linenum)
            elif "Test phase started at" in linetxt:
                t_starts.append(linenum)
    return (s_starts, t_starts)


fname='MSTlog_010_Rest_pre.txt'
fname_final=fname[:-4]
fname_final=fname_final + ".csv"
s_starts,t_starts=findlines(fname)
n_stim_per=64
study_len = n_stim_per*2
test_len = n_stim_per*3


study_data=pd.read_csv(fname,sep='\t',skiprows=s_starts[0],header=0,nrows=study_len)
test_data=pd.read_csv(fname,sep='\t',skiprows=t_starts[0],header=0,nrows=test_len, skipinitialspace=True)

#Write to CSV if you want
#test_data.to_csv(fname_final)

# Restrict dataframe to Lures that were answered correctly
#test_data_filter=test_data[test_data['Acc'] == 1]
#test_data_filter=test_data_filter[test_data['Cond'] == "L"]
```
